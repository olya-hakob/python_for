```python
def salary_update(old_salary,raise_percentage):
    return old_salary+(old_salary*raise_percentage)/100
new_salary = salary_update(100,10)
print(new_salary)
```

Կարող ենք գրել նաև ֆունկցիա, որը կհաշվի տվյալ list-ի միջին արժեքը՝

```python
def mean(x):
    x_sum = 0
    for item in x:
        x_sum = x_sum+item
#       x_sum += item
    return x_sum/len(x)
```

```python
def mean_shorter(x):
    return sum(x)/len(x)
```

```python
currency = '$'
old_salaries = [100,120,140,150,160,500,900,1500]
low_salaries = [] # դատարկ list որտեղ պահպանելու ենք ցածր աշխատավարձերը
middle_salaries = [] # դատարկ list որտեղ պահպանելու ենք միջին աշխատավարձերը
high_salaries = [] # դատարկ list որտեղ պահպանելու ենք բարձր աշխատավարձերը

for salary in old_salaries: 
    if salary < 500: 
        low_salaries.append(salary_update(salary,20)) #append նշանակում է կցել, նոր արժեքը կցվում է հատկացված list-ին
    elif 500<salary<1000:
        middle_salaries.append(salary_update(salary,10))
    else:
        high_salaries.append(salary_update(salary,5))
print(f'Ցածր աշխատավարձով աշխատողների թարմացված աշխատավարձի միջինը {mean(low_salaries)}{currency} է')
print(f'Ցածր աշխատավարձով աշխատողների թարմացված աշխատավարձի միջինը է {mean(middle_salaries)}{currency} է')
print(f'Ցածր աշխատավարձով աշխատողների թարմացված աշխատավարձի միջինը է {mean(high_salaries)}{currency} է')
```

```python
# Ուսումնասիրել հետևյալ կոդն ու նշումների միջոցով բացատրել, թե ինչ է այն անում --> մոդի հաշվարկը

list_of_numbers = [1,1,2,3,4,4,4,5,6,7,7] 
unique_numbers = list(set(list_of_numbers))
counter = [0]*len(unique_numbers)
for item in list_of_numbers:
    ind = unique_numbers.index(item)
    counter[ind] += 1
print(unique_numbers[counter.index(max(counter))] )  
```
